```python
original_dict = {}

with open('input.txt') as file:
    for f in file:
        key = f.split()[0].strip(':')
        value = f[6:].strip()
        
        original_dict[key] = value
        
for key, value in original_dict.items():
    try:
        original_dict[key] = int(value)
    except:
        pass

original_dict['humn'] = 'humn'
humn = 'humn'
solved_dict = {key: False for key, value in original_dict.items()}
humn_dep_dict = {key: False for key, value in original_dict.items()}
humn_dep_dict['humn'] = True
solved_dict['humn'] = True

while not solved_dict['root']:   
    for key, value in original_dict.items():
        if (type(value) == int and not solved_dict[key]):
            str_to_eval = f'{key} = {value}'
            exec(str_to_eval)
            solved_dict[key] = True
            print(f'Solved {key}')
        else:
            if not solved_dict[key]:
                str_to_eval = value
                first_term = value.split()[0]
                second_term = value.split()[2]
                try:
                    humn_encountered = False
                    eval(first_term)
                    eval(second_term)
                    
                    if humn_dep_dict[first_term] or humn_dep_dict[second_term]:
                        str_to_eval = str_to_eval.replace(first_term, f'({eval(first_term)})')
                        str_to_eval = str_to_eval.replace(second_term, f'({eval(second_term)})')
                        string_to_exec = f'{key} = "{str_to_eval}"'
                        print(f'Human encountered, therefore setting string {string_to_exec}')
                        exec(string_to_exec)
                        humn_dep_dict[key] = True
                    else:
                        print('Got here')
                        exec(f'{key} = {value}')

                    solved_dict[key] = True
                    print(f'Solved {key}')
                except:
                    print(f'Got error on {key}')
                    pass

# original_dict
```
