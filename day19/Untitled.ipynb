```python
current_blueprint = 0
blueprints = {}

with open('input.txt') as file:
    for f in file:
#         if f[0] == " ":
#             if counter == 0:
#                 ore_robot_ore = int(f.split()[4])
#             if counter == 1:
#                 clay_robot_ore = int(f.split()[4])
#             if counter == 2:
#                 obsidian_robot_ore, obsidian_robot_clay = int(f.split()[4]), int(f.split()[7])
#             if counter == 3:
#                 geode_robot_ore, geode_robot_obsidian = int(f.split()[4]), int(f.split()[7])
#                 blueprints[current_blueprint] = {
#                     'ore_robot_ore': ore_robot_ore,
#                     'clay_robot_ore': clay_robot_ore,
#                     'obsidian_robot_ore': obsidian_robot_ore,
#                     'obsidian_robot_clay': obsidian_robot_clay,
#                     'geode_robot_ore': geode_robot_ore,
#                     'geode_robot_obsidian': geode_robot_obsidian,
#                 }
#             counter += 1
#         else:
#             current_blueprint += 1
#             counter = 0
        current_blueprint = int(f.split()[1][:-1])
        ore_robot_ore = int(f.split()[6])
        clay_robot_ore = int(f.split()[12])
        obsidian_robot_ore = int(f.split()[18])
        obsidian_robot_clay = int(f.split()[21])
        geode_robot_ore = int(f.split()[27])
        geode_robot_obsidian = int(f.split()[30])
        blueprints[current_blueprint] = {
                        'ore_robot_ore': ore_robot_ore,
                        'clay_robot_ore': clay_robot_ore,
                        'obsidian_robot_ore': obsidian_robot_ore,
                        'obsidian_robot_clay': obsidian_robot_clay,
                        'geode_robot_ore': geode_robot_ore,
                        'geode_robot_obsidian': geode_robot_obsidian,
                    }

resource_types = ['ore', 'clay', 'obsidian', 'geode']
            
starting_robots = {
    'ore': 1,
    'clay': 0,
    'obsidian': 0,
    'geode': 0,
}
            
starting_resources = {
    'ore': 0,
    'clay': 0,
    'obsidian': 0,
    'geode': 0,
}


total_quality = 0


choices = [[0,1,2,3] for _ in range(10)]

def calculate_number_geodes_by_path(order, blueprint): 
    current_robots = starting_robots.copy()
    current_resources = starting_resources.copy()
    current_idx = 0
    length = len(order)
    order = list(order)
    
    added_0 = False
    added_1 = False
    added_2 = False
    added_3 = False
    
    max_ore_robots = max(blueprint['ore_robot_ore'], blueprint['clay_robot_ore'], blueprint['obsidian_robot_ore'], blueprint['geode_robot_ore'])
    max_clay_robots = blueprint['obsidian_robot_clay']
    max_obsidian_robots = blueprint['geode_robot_obsidian']
    
    for turn in range(1,33):
#         print(turn)
#         print(current_robots)
#         print(current_resources)
        prebuild_current_robots = current_robots.copy()
        premine_current_resources = current_resources.copy()
        built_this_turn = False
        
        # Mine
        for resource in resource_types:
            current_resources[resource] += prebuild_current_robots[resource]
        
        # Build
        
        if (premine_current_resources['ore'] >= blueprint['geode_robot_ore']) & ((premine_current_resources['obsidian'] >= blueprint['geode_robot_obsidian'])):
            if (current_idx == length):
                if (len(order) <= 24) & (~added_3) & (turn <= 31):
                    backlog.append(order + [3])
#                     print(f'Adding {order + [3]} to backlog')
                    added_3 = True
                    continue
            elif (order[current_idx] == 3)  & (~built_this_turn):
                current_robots['geode'] += 1
                current_resources['ore'] -= blueprint['geode_robot_ore']
                current_resources['obsidian'] -= blueprint['geode_robot_obsidian']
                current_idx += 1
                built_this_turn = True
                continue
        
        if (premine_current_resources['ore'] >= blueprint['ore_robot_ore']):
            if (current_idx == length):
                if (len(order) <= 24) & (~added_0) & (turn <= 29) & (current_robots['ore'] * (29-turn) + premine_current_resources['ore'] <= max_ore_robots * (29-turn)):
                    backlog.append(order + [0])
#                     print(f'Adding {order + [0]} to backlog')
                    added_0 = True
            elif (order[current_idx] == 0) & (~built_this_turn):
                current_robots['ore'] += 1
                current_resources['ore'] -= blueprint['ore_robot_ore']
                current_idx += 1
                built_this_turn = True
#                 print('Building ore robot')

        if (premine_current_resources['ore'] >= blueprint['clay_robot_ore']):
            if (current_idx == length):
                if (len(order) <= 24) & (~added_1) & (turn <= 27) & (current_robots['clay'] * (27-turn) + premine_current_resources['clay'] <= max_clay_robots * (27-turn)):
                    backlog.append(order + [1])
#                     print(f'Adding {order + [1]} to backlog')
                    added_1 = True
            elif (order[current_idx] == 1)  & (~built_this_turn):
                current_robots['clay'] += 1
                current_resources['ore'] -= blueprint['clay_robot_ore']
                current_idx += 1
                built_this_turn = True
#                 print('Building clay robot')

        if (premine_current_resources['ore'] >= blueprint['obsidian_robot_ore']) & (premine_current_resources['clay'] >= blueprint['obsidian_robot_clay']):
            if (current_idx == length):
                if (len(order) <= 24) & (~added_2) & (turn <= 29) & (current_robots['obsidian'] * (29-turn) + premine_current_resources['obsidian'] <= max_obsidian_robots * (29-turn)):
                    backlog.append(order + [2])
#                     print(f'Adding {order + [2]} to backlog')
                    added_2 = True
            elif (order[current_idx] == 2)  & (~built_this_turn):
                current_robots['obsidian'] += 1
                current_resources['ore'] -= blueprint['obsidian_robot_ore']
                current_resources['clay'] -= blueprint['obsidian_robot_clay']
                current_idx += 1
                built_this_turn = True
#                 print('Building obsidian robot')


        

            
        
    return current_resources['geode']


# blueprints = {1: {'ore_robot_ore': 4,
#   'clay_robot_ore': 2,
#   'obsidian_robot_ore': 3,
#   'obsidian_robot_clay': 14,
#   'geode_robot_ore': 2,
#   'geode_robot_obsidian': 7},
#  2: {'ore_robot_ore': 2,
#   'clay_robot_ore': 3,
#   'obsidian_robot_ore': 3,
#   'obsidian_robot_clay': 8,
#   'geode_robot_ore': 3,
#   'geode_robot_obsidian': 12}}

# for idx, blueprint in blueprints.items():
for idx, blueprint in {i:k for i, k in blueprints.items() if i <= 3}.items():
    print(f'Starting for idx {idx}')
    most_geodes_so_far = 0

#     # Create a backlog of options to explore
#     backlog = [
#         [(0,0,0)] # For each turn, this is the number of ore robots, number of clay robots, number of obsidian robots to make
#     ]
    
    
#     while len(backlog) > 0:
#         path = backlog.pop(0)
#         geodes = calculate_number_geodes_by_path(path, blueprint)
#         if geodes > most_geodes_so_far:
#             logger.debug(f'Found path with {geodes} geodes in blueprint {idx}')
#             most_geodes_so_far = geodes
#         print(f'Number of paths in backlog now: {len(backlog)}')
    
#     for order in tqdm_notebook(product([0,1], [0,1], *choices, [3])):
# #         print(order)
#         geodes = calculate_number_geodes_by_path(order, blueprint)
#         if geodes > most_geodes_so_far:
#             print(f'Found path with {geodes} geodes in blueprint {idx} with order {order}')
#             most_geodes_so_far = geodes
            
    backlog = [[0], [1]]
    
    while len(backlog) > 0:
        order = backlog.pop(0)
        geodes = calculate_number_geodes_by_path(order, blueprint)
        if geodes > most_geodes_so_far:
            print(f'Found path with {geodes} geodes in blueprint {idx} with order {order}')
            most_geodes_so_far = geodes
#         print(f'Backlog: {len(backlog)}')
    
    
    total_quality += idx * most_geodes_so_far

print(total_quality)
```
