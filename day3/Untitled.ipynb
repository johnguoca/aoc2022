```python
# Greedily go from both ends, storing what's found in hashtable. Once we find one that's already in hashtable, that's the answer
total = 0
char = ""
with open('input.txt') as file:
    for f in file:
#         print(f)
        first_d = {}
        second_d = {}
        current = 0
        while True:
            start = f[current]
            end = f[-(current + 1)]
            first_d[start] = first_d.get(start, 0) + 1
            second_d[end] = second_d.get(end, 0) + 1
            if second_d.get(start, 0) > 0:
                char = start
                break
            if first_d.get(end, 0) > 0:
                char = end
                break
            current += 1
        # Get the priority
        # ASCII of a-z is 97-122
        # ASCII of A-Z is 65-90
#         print(char)
        if ord(char) < 91:
            priority = ord(char) - 38
        else:
            priority = ord(char) - 96
#         print(f'Priority: {priority}')

        total += priority
print(total)
```
