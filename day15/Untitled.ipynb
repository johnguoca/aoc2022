```python
# possible_points = {} # Point: which circle it showed up in
sensor_coords = []
beacon_coords = []
counter = 0

coord_min = 0
# coord_max = 20
coord_max = 4000000

possible_points = []

with open('input.txt') as file:
    for f in file:
        counter += 1
        print(f)
        x_sensor = int(f.split()[2][2:-1])
        y_sensor = int(f.split()[3][2:-1])
        x_beacon = int(f.split()[-2][2:-1])
        y_beacon = int(f.split()[-1][2:])
        
        distance = abs(x_sensor - x_beacon) + abs(y_sensor - y_beacon)
        
        distance_plus_one = distance + 1
        beacon_coords.append([x_beacon, y_beacon])
        sensor_coords.append([x_sensor, y_sensor])
        
        # Get all the points that are distance + 1 away
        for x in range(-distance_plus_one, distance_plus_one + 1):
            x_possible = x_sensor + x
            y_delta = distance_plus_one - abs(x)
            
            if x_possible >= 0 and x_possible <= coord_max and (y_sensor - y_delta >= 0) and (y_sensor - y_delta <= coord_max):
#                 possible_points[(x_possible, y_sensor - y_delta)] = possible_points.get((x_possible, y_sensor - y_delta), []) + [counter]
                possible_points.append((x_possible, y_sensor - y_delta))
            if x_possible >= 0 and x_possible <= coord_max and (y_sensor + y_delta >= 0) and (y_sensor + y_delta <= coord_max):
                possible_points.append((x_possible, y_sensor + y_delta))
#                 possible_points[(x_possible, y_sensor + y_delta)] = possible_points.get((x_possible, y_sensor + y_delta), []) + [counter]
            
# possible_points         
```

```python
def manhattan_dist(p1, p2):
    return abs(p1[0]- p2[0]) + abs(p1[1] - p2[1])

for point in tqdm_notebook(possible_points):
    current_check = True
    for ps, pb in zip(sensor_coords, beacon_coords):
        if manhattan_dist(ps, point) <= manhattan_dist(ps, pb):
            current_check = False
    if not current_check:
        continue
    else:
        print(f'WE FOUND IT: {point}')
```
