```python
import pandas as pd
import numpy as np
from queue import Queue
```

```python
cubes = []
cube_already_there = []
surface_area = 0

with open('input.txt') as file:
    for f in file:
        x,y,z = f.strip().split(',')
        cubes.append((int(x), int(y), int(z)))
        
max_x, max_y, max_z = max([cube[0] for cube in cubes]), max([cube[1] for cube in cubes]), max([cube[2] for cube in cubes])
min_x, min_y, min_z = min([cube[0] for cube in cubes]), min([cube[1] for cube in cubes]), min([cube[2] for cube in cubes])

full_cube = np.zeros((max_x + 1, max_y + 1, max_z + 1))
already_checked = np.zeros((max_x + 1, max_y + 1, max_z + 1))

for idx, (x,y,z) in enumerate(cubes):
#     print(f'Putting cube at {x, y, z}')
    
    if full_cube[x,y,z] == 1:
        cube_already_there.append(idx)
        print(x,y,z)
    full_cube[x,y,z] = 1

for idx, (x,y,z) in enumerate(cubes):
    if idx in cube_already_there:
        continue
    
    if (x - 1 == -1):
        surface_area += 1
    elif full_cube[x-1, y, z] != 1:
        surface_area += 1
        
    if (x == max_x):
        surface_area += 1
    elif full_cube[x+1, y, z] != 1:
        surface_area += 1

    if (y - 1 == -1):
        surface_area += 1
    elif full_cube[x, y - 1, z] != 1:
        surface_area += 1
        
    if (y == max_y):
        surface_area += 1
    elif full_cube[x, y + 1, z] != 1:
        surface_area += 1
        
    if (z - 1 == -1):
        surface_area += 1
    elif full_cube[x, y, z - 1] != 1:
        surface_area += 1
        
    if (z == max_z):
        surface_area += 1
    elif full_cube[x, y, z + 1] != 1:
        surface_area += 1
        
print(surface_area)


# For each cube, try to find a way out
bounded_sides = 0    

for x_start in range(full_cube.shape[0]):
    for y_start in range(full_cube.shape[1]):
        for z_start in range(full_cube.shape[2]):
#             print(f'Starting at {x_start, y_start, z_start}')
            inner_sides_found = 0
            backlog = []
            trapped = True
            if (already_checked[x_start,y_start,z_start] == 1) or (full_cube[x_start,y_start,z_start]):
                trapped = False
                continue
            else:
                backlog.append((x_start,y_start,z_start))
                while len(backlog) > 0:
                    x,y,z = backlog.pop()
                    if already_checked[x,y,z] == 1:
                        continue
                    else:
                        already_checked[x,y,z] = 1
#                     print(f'Checking {x,y,z}')
                    if (x == 0):
                        trapped = False
                    elif already_checked[x-1, y, z] == 1:
                        pass
                    elif full_cube[x-1, y, z] == 1:
                        inner_sides_found += 1
                    else:
                        backlog.append((x-1, y, z))
                    
                    if (x == max_x):
                        trapped = False
                    elif already_checked[x+1, y, z] == 1:
                        pass
                    elif full_cube[x + 1, y, z] == 1:
                        inner_sides_found += 1
                    else:
                        backlog.append((x + 1, y, z))
                    
                    
                    if (y == 0):
                        trapped = False
                    elif already_checked[x, y - 1, z] == 1:
                        pass
                    elif full_cube[x, y - 1, z] == 1:
                        inner_sides_found += 1
                    else:
                        backlog.append((x, y - 1, z))
                    
                    if (y == max_y):
                        trapped = False
                    elif already_checked[x, y + 1, z] == 1:
                        pass
                    elif full_cube[x, y + 1, z] == 1:
                        inner_sides_found += 1
                    else:
                        backlog.append((x, y + 1, z))
                        
                    if (z == 0):
                        trapped = False
                    elif already_checked[x, y, z - 1] == 1:
                        pass
                    elif full_cube[x, y, z - 1] == 1:
                        inner_sides_found += 1
                    else:
                        backlog.append((x, y, z - 1))
                    
                    if (z == max_z):
                        trapped = False
                    elif already_checked[x, y, z + 1] == 1:
                        pass
                    elif full_cube[x, y, z + 1] == 1:
                        inner_sides_found += 1
                    else:
                        backlog.append((x, y, z + 1))
                if trapped:
                    bounded_sides += inner_sides_found
                    print(f'Adding {inner_sides_found} to inner sides found at {x_start, y_start, z_start}')
                    
print(surface_area - bounded_sides)
```
