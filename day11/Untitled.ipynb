```python
import copy
dict_to_use = copy.deepcopy(input_dict)

divisors = [i['divisible'] for i in dict_to_use.values()]  

class Item(object):
    def __init__(self, initial_value, divisors, divisibility_grid):
        self.divisibility_grid = divisibility_grid
        for d in divisors:
            self.divisibility_grid[d] = initial_value % d
    
for key, value in dict_to_use.items():
    value['item_list'] = [Item(initial_value, divisors, {}) for initial_value in value['items']]
    
inspected_dict = {i: 0 for i in dict_to_use.keys()}

rounds = 10000

for round_number in range(rounds):
    print(f'ROUND NUMBER {round_number}')
    for monkey_number, monkey in dict_to_use.items():
        items = monkey['item_list']
        operator, value = monkey['op'].strip().split()
        if value != 'self':
            value = int(value)
        divisible = monkey['divisible']
        true_monkey = monkey['true']
        false_monkey = monkey['false']
        for _ in range(len(items)):
            inspected_dict[monkey_number] += 1
            item_to_inspect = items.pop()
            div_grid = item_to_inspect.divisibility_grid
            if operator == '+':
                for divisor, remainder in div_grid.items():
                    div_grid[divisor] = ((remainder % divisor) + (value % divisor)) % divisor
            else:
                if value == 'self':
                    for divisor, remainder in div_grid.items():
                        div_grid[divisor] = (remainder * remainder) % divisor
                else:
                    for divisor, remainder in div_grid.items():
                        div_grid[divisor] = ((remainder % divisor) * (value % divisor)) % divisor
            if div_grid[divisible] == 0:
                dict_to_use[true_monkey]['item_list'].append(item_to_inspect)
            else:
                dict_to_use[false_monkey]['item_list'].append(item_to_inspect)
    print(inspected_dict)

monkey_business = sorted(inspected_dict.values())[-1] * sorted(inspected_dict.values())[-2]
print(monkey_business)
```
