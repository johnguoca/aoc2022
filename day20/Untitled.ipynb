```python
import numpy as np
from tqdm import tqdm_notebook
```

```python
li = []

with open('input.txt') as file:
    for f in file:
        li.append(int(f.strip()))

li = [811589153 * i for i in li]
pointers = [i for i in range(len(li))]
```

```python
li_to_r = [(i, False) for i in li]
for _ in tqdm_notebook(range(10)):
    li_to_r = [(i, False) for i, k in li_to_r]
#     print(li_to_r)
    
    current_pointer = 0
    while current_pointer <= len(li_to_r) - 1:
        idx_to_find = pointers.index(current_pointer)
        p = li_to_r.pop(idx_to_find)
        pointer_p = pointers.pop(idx_to_find)

        new_index = (idx_to_find + p[0]) % len(li_to_r)
        if new_index == 0 and p[0] != 0:
            new_index = len(li_to_r)
        li_to_r.insert(new_index, (p[0], True))
        pointers.insert(new_index, pointer_p)
#         print(current_pointer)
#         print(p)
#         print(new_index)
        current_pointer += 1
#         print(li_to_r)
    print(li_to_r)
```
